### Problem Statement:

The banking churn prediction dataset (uploaded in the same folder) contains different attributes of the customers of the bank and whether they have churned or not.
Churning means closing the relationship with the bank. Following are the descriptions of the columns in the dataset.

- customer_id: Unique identifier for each customer.
- vintage: The duration of the customer's relationship with the company.
- age: Age of the customer.
- gender: Gender of the customer.
- dependents: Number of dependents the customer has.
- occupation: The occupation of the customer.
- city: City in which the customer is located.
- customer_nw_category: Net worth category of the customer.
- branch_code: Code identifying the branch associated with the customer.
- current_balance: Current balance in the customer's account.
- previous_month_end_balance: Account balance at the end of the previous month.
- average_monthly_balance_prevQ: Average monthly balance in the previous quarter.
- average_monthly_balance_prevQ2: Average monthly balance in the second previous quarter.
- current_month_credit: Credit amount in the current month.
- previous_month_credit: Credit amount in the previous month.
- current_month_debit: Debit amount in the current month.
- previous_month_debit: Debit amount in the previous month.
- current_month_balance: Account balance in the current month.
- previous_month_balance: Account balance in the previous month.
- churn: The target variable indicating whether the customer has churned (1 for churned, 0 for not churned).
- last_transaction: Timestamp of the customer's last transaction.


Your task is to perform exploratory data analysis (EDA) on Banking Churn Prediction dataset. 

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('Banking_churn_prediction.csv')
data_five=data.head(5)  # Display the first 5 rows of the DataFrame
print(data.head())  # Display the first few rows of the DataFrame
print(data.info())#gives info of all columns,dtype,counts
print(data.describe())#provides statistical summary of numerical columns-mean,mode,25%//
print(data.isnull().sum())#checks for null values in each column
print(data.isna().sum())#both same
print(data.shape)  # Returns the shape of the DataFrame (rows, columns)
print(data.columns)  # Returns the column names of the DataFrame
print(data.dtypes)  # Returns the data types of each column
print(data.nunique())  # Returns the number of unique values in each column
# Check Duplicates	
print(data.duplicated().sum())

null_counts = data.isnull().sum()
print( null_counts[null_counts > 0])  # Displays columns with null values
null_columns=null_counts[null_counts > 0]
for col, count in null_columns.items():
    print(f"{col}: {count} missing values")# to iterate with that

# Value counts for categorical columns
for col in data.select_dtypes(include='object').columns:
    print(f"\n{col} value counts:\n", data[col].value_counts())

# ## Convert last_transaction to datetime
'''errors='coerce'--> If there are any values that cannot be converted to datetime,
 they will be set to NaT (Not a Time) instead of raising an error.'''

data['last_transaction'] = pd.to_datetime(data['last_transaction'], errors='coerce')
print(data['last_transaction'].head())  # Display the first few rows of the 'last_transaction' column
# Calculate days since last transaction
latest_date = data['last_transaction'].max()
data['days_since_last_transaction'] = (latest_date - data['last_transaction']).dt.days
print(data['days_since_last_transaction'].head())  


# Histograms fOr numeric column
data.hist(bins=30, figsize=(40,20))
plt.show()

# Boxplots to check outliers for all columns quantitative/numerical
for col in data.select_dtypes(include='number').columns:
    sns.boxplot(y=col, data=data)
    plt.title(col)
    plt.show()

# heatmap showing where missing data (NaN) exists. White = missing, Black = not missing (default)
sns.heatmap(data.isnull(), cbar=False)
plt.show()

#Univariate analysis
# Count plot for categorical column 'Churn'
sns.countplot(x='churn', data=data)
plt.title('Count of Churn')
plt.xlabel('churn')
plt.ylabel('Count')
plt.show()
# Distribution of numerical column 'Age'
sns.histplot(data['age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
# Boxplot for numerical column 'Age' by 'Churn'
sns.boxplot(x='churn', y='age', data=data)
plt.title('Age by Churn')
plt.xlabel('Churn')
plt.ylabel('age')
plt.show()


'''If inplace=True, the changes are made to the original DataFrame and nothing is returned.
If inplace=False (default), the method returns a new DataFrame with the changes, 
and the original DataFrame remains unchanged.'''
# Replace missing values for correlation analysis error
'''Imputed using mode for categorical and median for numerical.'''
for col in data.columns:
    if data[col].dtype == 'object':
        # Replace with mode for categorical columns
        data[col].fillna(data[col].mode()[0], inplace=False)
    else:
        # Replace with median for numerical columns
        data[col].fillna(data[col].median(), inplace=False)
print(data.isnull().sum()) 
print(data['churn'].isnull().sum()) 

# Correlation heatmap for numerical columns only
# #only the columns from data which have numeric data types (such as int, float, etc.).
numeric_data = data.select_dtypes(include=[np.number])
#Convert 'churn' to numeric 
data['churn'] = pd.to_numeric(data['churn'], errors='coerce')
plt.figure(figsize=(12, 8))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()

##### skewness
'''in pandas calculates the skewness of a numerical column, which measures the asymmetry of the data
 distribution.

Positive skew: Tail is longer on the right side (more values are concentrated on the left).
Negative skew: Tail is longer on the left side (more values are concentrated on the right).
Zero skew: Data is symmetric (like a normal distribution).'''
# # Calculate skewness for numerical columns
''' the skewness for all numeric columns.'''
skewness = data.select_dtypes(include=[np.number]).apply(lambda x: x.skew()).sort_values(ascending=False)
# Print skewness values
print(skewness)
''' Why Do We Log Transform?
To reduce the skewness of right-skewed data
Make the data more normally distributed
Improve performance of statistical models and visualizations
Handle large numerical ranges more smoothly
When to apply:
Only to positive-valued columns (no zero or negative)
Only to highly skewed features (skewness > 1 is a good threshold)
When Not to Use Log Transform
Situation	Reason
Column has 0 or negative values and not log1p	Will cause math errors
Column is already normal	No benefit
Categorical data	Log doesnt apply

Transform	Use Case	Formula
np.log(x)	Positive numbers > 0	log(x)
np.log1p(x)	Includes 0	log(x + 1)
np.sqrt(x)	Mild skew, safe for 0	sqrt(x)
We typically use np.log1p() because it handles 0 safely.
'''
#Apply log transform for columns 
for col in skewness.index:
    if skewness[col] > 1:  # Only highly skewed
        if (data[col] >= 0).all():  # Ensure no negative numbers
            data[col + '_log'] = np.log1p(data[col])
print(data['current_balance'].skew())# 99.030683
data['current_balance_log'] = np.log1p(data['current_balance'])
print(data['current_balance_log'].skew())
'''Visual Comparison (Before vs After)'''

# Before
sns.histplot(data['current_balance'], kde=True)
plt.title('Before Log Transform')
plt.show()

# After
sns.histplot(data['current_balance_log'], kde=True)
plt.title('After Log Transform')
plt.show()


# # Plot skewness for numerical columns
''' function gives you three detailed skewness measures for one column,
showing how much and in what direction the data is skewed.'''
def skewness_measure(x):
    mean = x.mean()
    std = x.std()
    q1, q2, q3 = x.quantile([0.25, 0.5, 0.75]).values
    mode = x.value_counts().head(1).index.values[0]

    sk1 = (mean - mode)/std
    sk2 = 3*(mean - q2)/std
    skb = (q1+q3 - 2*q2)/(q3-q1)
    IQR = q3-q1
    print(f"Mean = {mean:.2f}, Median = {q2:.2f}, Mode = {mode:.2f}, Std Dev = {std:.2f}, IQR = {IQR:.2f}")
    print(f"Pearson's first coefficient of skewness = {sk1:.2f},\nPearson's second coefficient of skewness = {sk2:.2f},\nBowley's coefficient of skewness = {skb:.2f}")
  
skewness_measure(data['current_month_credit'])

# ##This line removes outliers from your DataFrame using the IQR (Interquartile Range) method.
# '''This line removes **outliers** from your DataFrame using the **IQR (Interquartile Range) method**.
# for Cleaned dataset
# ### Explanation:
# '''Both Z-score and IQR methods are commonly used for outlier removal, but the best choice depends on your data:

# IQR Method
# Best for: Data that is not normally distributed or has skewness.
# How it works: Removes values outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR].
# Pros: Robust to non-normal data and skewed distributions.
# Cons: May miss outliers in the tails if data is normal.
# Z-score Method
# Best for: Data that is normally distributed (bell-shaped).
# How it works: Removes values with Z-score above a threshold (commonly |z| > 3).
# Pros: Works well for symmetric, normal data.
# Cons: Sensitive to mean and standard deviation; not robust for skewed data.
# Summary:

# Use IQR for skewed or non-normal data (most real-world cases).
# Use Z-score for data that is approximately normal.
# In practice, IQR is usually safer for general datasets.'''
# data_no_outliers_iqr = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
# ```
# - `Q1` is the 25th percentile (lower quartile) for each column.
# - `Q3` is the 75th percentile (upper quartile) for each column.
# - `IQR = Q3 - Q1` is the interquartile range.
# - The expression `(data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))` creates a boolean DataFrame marking values outside the "normal" range (outliers).
# - `.any(axis=1)` checks if **any column** in a row is an outlier.
# - `~` negates the boolean mask, so only rows **without any outlier values** are kept.
# `data_no_outliers_iqr` is a DataFrame with all rows containing at least one outlier removed.
#does not work for categorical/object columnsit only works for numeric columns.'''
numeric_cols = data.select_dtypes(include=[np.number]).columns
Q1 = data[numeric_cols].quantile(0.25)
Q3 = data[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
data_no_outliers_iqr = data[~((data[numeric_cols] < (Q1 - 1.5 * IQR)) | (data[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
print("Data without outliers based on IQR:", data_no_outliers_iqr.head())
#check if the shape of the data has changed after removing outliers
print("Original data shape:", data.shape)
print("Data shape after outlier removal:", data_no_outliers_iqr.shape)

# Optionally, check summary statistics
print("Original data stats:\n", data.describe())
print("Data without outliers stats:\n", data_no_outliers_iqr.describe())
'''from scipy import stats

zscore_data = data.select_dtypes(include=[np.number]).copy()
z_scores = np.abs(stats.zscore(zscore_data.dropna()))  # dropna to handle NaNs

# Filter rows where all z-scores < 3 (commonly used threshold)
data_no_outliers_zscore = data[(z_scores < 3).all(axis=1)]'''


data['gender'].value_counts().plot(kind='bar')
plt.show()

print(data.columns)

#Bivariate analysis
'''We examine and understand the relationship among two variables
 relationships between churn and other variable
 '''

print(data.select_dtypes(include='object').columns)      # Categorical columns
print(data.select_dtypes(include=[np.number]).columns)   # Numerical columns
print(data_five['last_transaction']) #object type column
print(pd.to_datetime(data_five['last_transaction'])) ## Convert 'last_transaction' to datetime-DATETIME TYPE

#Value Distribution by Churn
sns.histplot(data=data, x='age', hue='churn', kde=True)
plt.title('Age Distribution by Churn')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

#Numerical/quantative vs Target --q-->c
#box plot-These plots show if customers who churned have different balance/age patterns than those who didnt.
numerical_cols = ['age', 'vintage', 'current_balance', 'average_monthly_balance_prevQ',
                  'average_monthly_balance_prevQ2', 'current_month_credit',
                  'current_month_debit', 'current_month_balance']

for col in numerical_cols:
    sns.boxplot(x='churn', y=col, data=data)
    plt.title(f'{col} by Churn')
    plt.xlabel('Churn')
    plt.ylabel(col)
    plt.show()

#Numerical vs Numerical q-q--scatter plot-Use for numeric variable pair comparison with churn as hue.
sns.scatterplot(x='current_month_credit', y='current_month_debit', hue='churn', data=data)
plt.title('Current Month Credit vs Debit by Churn')
plt.show()

# #Categorical vs Target --c-->c    --Do certain genders or occupations churn more?--its barplot/can use countplot same
categorical_cols = ['gender', 'occupation', 'customer_nw_category']

for col in categorical_cols:
    sns.countplot(x=col, hue='churn', data=data)
    plt.title(f'{col} vs Churn')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Churn Rate by Categorical Variable-churn percentage within each group.
# Churn rate by occupation
churn_rate = data.groupby('occupation')['churn'].mean().sort_values(ascending=False)
churn_rate.plot(kind='bar', title='Churn Rate by Occupation')
plt.ylabel('Churn Rate')
plt.show()

#Heatmap of Feature Correlations
corr = data.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr[['churn']].sort_values(by='churn', ascending=False), annot=True, cmap='coolwarm')
plt.title('Correlation of Features with Churn')
plt.show()
````
